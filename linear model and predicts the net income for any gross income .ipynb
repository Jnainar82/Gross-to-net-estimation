```python
import matplotlib.pyplot as plt

gross_1= 2000
net_1=1477.66
gross_2 =2100
net_2 = 1511.19

a = (net_2-net_1)/(gross_2-gross_1)
#print(a)

b= net_1-a*gross_1
#print(b)

def predict_net_income(gross):
  return a*gross+b

gross_values = list(range(1800, 3001, 100))  # from 1800 to 3000, step by 100
net_values = [predict_net_income(g) for g in gross_values]

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(gross_values, net_values, marker='o', linestyle='-', color='blue', label='Predicted Net Income')
plt.scatter([gross_1, gross_2], [net_1, net_2], color='red', label='Original Data Points')
plt.axvline(2800, color='green', linestyle='--', label='Gross = 2800')

# Annotate prediction
predicted_2800 = predict_net_income(2800)
plt.scatter(2800, predicted_2800, color='purple')
plt.text(2800, predicted_2800 + 10, f'{predicted_2800:.2f}', ha='center', color='purple')

# Labels and legend
plt.title("Gross vs Net Income Prediction")
plt.xlabel("Gross Income")
plt.ylabel("Net Income")
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()


#print(predict_net_income(2000))

gross_income = 2800
net_income = predict_net_income(gross_income)

print(f"Estimated net income for gross {gross_income} is: {net_income:.2f}")
```
